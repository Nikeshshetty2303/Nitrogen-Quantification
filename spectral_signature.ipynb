```python
from osgeo import gdal
import matplotlib.pyplot as plt

# Load the hyperspectral image using GDAL
image_path = 'raw_file.tif'
dataset = gdal.Open(image_path)

# Read the image data
img = dataset.ReadAsArray()

# Get the spectral signature of a specific pixel by specifying both band and pixel number
pixel_number = 1000  # Replace with your desired pixel number
spectral_signature = img[:, pixel_number]

# Plot the spectral signature
plt.figure(figsize=(10, 6))
plt.plot(spectral_signature, marker='o', linestyle='-')
plt.title('Spectral Signature of Hyperspectral Image')
plt.xlabel('Band Number')
plt.ylabel('Reflectance Value')
plt.grid(True)
plt.show()

# Close the dataset
dataset = None
```
